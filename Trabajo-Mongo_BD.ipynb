# Funciones CRUD básicas

```python
#Insertar un estudiante
def insertar_estudiante(nombre, edad, carrera):
    estudiante = {"nombre": nombre, "edad": edad, "carrera": carrera}
    estudiantes.insert_one(estudiante)
    print(f"Estudiante {nombre} insertado.")
```

```python
# Listar estudiantes
def listar_estudiantes():
    print("Listado de estudiantes:")
    for est in estudiantes.find():
        print(est)
```

```python
# Actualizar edad de un estudiante
def actualizar_edad(nombre, nueva_edad):
    estudiantes.update_one({"nombre": nombre}, {"$set": {"edad": nueva_edad}})
    print(f"Edad de {nombre} actualizada.")
```

```python
# Eliminar un estudiante
def eliminar_estudiante(nombre):
    estudiantes.delete_one({"nombre": nombre})
    print(f"Estudiante {nombre} eliminado.")
```

```python
def menu():
    while True:
        print("\n--- MENÚ DE ESTUDIANTES ---")
        print("1. Insertar estudiante")
        print("2. Listar estudiantes")
        print("3. Actualizar edad")
        print("4. Eliminar estudiante")
        print("5. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            nombre = input("Nombre: ")
            edad = int(input("Edad: "))
            carrera = input("Carrera: ")
            insertar_estudiante(nombre, edad, carrera)

        elif opcion == "2":
            listar_estudiantes()

        elif opcion == "3":
            nombre = input("Nombre del estudiante: ")
            nueva_edad = int(input("Nueva edad: "))
            actualizar_edad(nombre, nueva_edad)

        elif opcion == "4":
            nombre = input("Nombre del estudiante: ")
            eliminar_estudiante(nombre)

        elif opcion == "5":
            print("Saliendo...")
            break

        else:
            print("Opción inválida")
```
